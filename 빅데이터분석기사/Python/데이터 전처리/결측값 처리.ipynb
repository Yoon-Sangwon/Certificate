```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.describe(include = 'all')
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# NaN 값이 500개 이상인 열을 모두 삭제 - deck 열(891개 중 688개의 NaN 값)
df.dropna(axis=1, thresh=500, inplace=True)
print(df.shape)
df.isnull().sum()
```

```python
df2 = df.dropna(subset=['age', 'embarked'])
df2.shape
```

```python
# 앞쪽으로 채우기
df[['age', 'embarked']].fillna(method ='ffill').head()
```

```python
# 뒤쪽으로 채우기
df[['age', 'embarked']].fillna(method ='bfill').head()
```
